```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("hearing_test.csv")
df.head()
```

```python
# Check for null values
null_values = df.isnull().sum()
print("Null Values:\n", null_values)
```

```python
#Check distribution of dataset
df.describe()
```

```python
# Univariate Analysis

# Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.show()

# Physical Score Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['physical_score'], bins=10, kde=True)
plt.title('Physical Score Distribution')
plt.show()

# Test Result Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='test_result', data=df)
plt.title('Test Result Distribution')
plt.show()
```

```python
# Bivariate Analysis

# Age vs. Physical Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='physical_score', data=df)
plt.title('Age vs. Physical Score')
plt.show()

# Age vs. Test Result
plt.figure(figsize=(6, 4))
sns.barplot(x='test_result', y='age', data=df)
plt.title('Age vs. Test Result')
plt.show()

# Physical Score vs. Test Result
plt.figure(figsize=(6, 4))
sns.barplot(x='test_result', y='physical_score', data=df)
plt.title('Physical Score vs. Test Result')
plt.show()
```

```python
# Multivariate Analysis

# Age, Physical Score, and Test Result
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='test_result')
plt.suptitle('Pair Plot of Age, Physical Score, and Test Result')
plt.show()
```
